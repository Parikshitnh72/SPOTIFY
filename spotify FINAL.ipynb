```python

print("Column names in DataFrame:", tracks.columns)


tracks.columns = tracks.columns.str.strip()


print(tracks.head())


tracks['release_date'] = pd.to_datetime(tracks['release_date'], errors='coerce')


tracks.dropna(subset=['release_date'], inplace=True)


tracks.set_index('release_date', inplace=True)


print(tracks.head())
```

```python
tracks[['artists']].iloc[18]
```

```python
tracks['duration'] = tracks['duration_ms'].apply (lambda x : round(x/1000))
tracks.drop('duration_ms', inplace = True, axis=1)
tracks.duration.head()
```

Correlation Heatmap between Variable using Pearson correlation method

```python
# Identify and keep only numeric columns
numeric_tracks = tracks.select_dtypes(include=[np.number])

# Calculate the correlation matrix
cm = numeric_tracks.corr(method='pearson')

# Plot the heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(cm, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
plt.show()
```

Considering 0.4% of the total dataset to create Regression plots

```python
sam = tracks.sample(int(0.004 * len(tracks)))
len(sam)
```

Regression plot - Correlation between Loudness and Energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sam, y='loudness', x='energy', color='c').set(title='Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sam, y='popularity', x='acousticness', color='b').set(title='Popularity vs Acousticness Correlation')
```

```python
# Ensure 'release_date' is in the index
tracks['dates'] = tracks.index.get_level_values('release_date')

# Convert the 'dates' column to datetime, coercing errors
tracks['dates'] = pd.to_datetime(tracks['dates'], errors='coerce')

# Drop rows where 'dates' could not be converted
tracks.dropna(subset=['dates'], inplace=True)

# Extract the year from the 'dates' column
years = tracks['dates'].dt.year

# Print the first few rows to check the result
print(tracks.head())
```

```python
tracks.head()
```

Distribution plot - Visualize total number of songs on Spotify since 1992

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year')
```

```python
total_dr = tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title='Years vs Duration')
plt.xticks(rotation=90)
```
